```r
fit = lm(wage~poly(age, 4), data = Wage)
```

```r
plot(fit)
```

```r
summary(fit)
```

```r
agelims = range(age)
age.grid = seq(from=agelims[1], to = agelims[2])
```

```r
age.grid
```

```r
preds = predict(fit, newdata = list(age=age.grid), se = TRUE)
```

```r
preds
```

```r
se.bands = cbind(preds$fit+2*preds$se.fit, preds$fit-2*preds$se.fit)
```

```r
plot(age, wage, col="darkgrey")
lines(age.grid, preds$fit, lwd = 2, col = 'blue')
matlines(age.grid, se.bands, col = 'red', lty=2)
```

But this works only in linear and single variable model. 

In more general, we would use 'anova()'.

```r
fita = lm(wage~education, data = Wage)
fitb = lm(wage~education+age, data = Wage)
fitc = lm(wage~education+poly(age, 2), data = Wage)
fitd = lm(wage~education+poly(age, 3), data = Wage)

anova(fita, fitb, fitc, fitd)
```

```r
fit = glm(I(wage>250)~poly(age, 3), data = Wage, family = binomial)
```

```r
summary(fit)
```

```r
preds = predict(fit, list(age=age.grid), se = TRUE)
```

```r
preds
```

```r
se.bands = preds$fit + cbind(fit = 0, lower = -2*preds$se.fit,
                             upper = 2*preds$se.fit)
```

```r
se.bands[1:5, ]
```

We have done the calculation in logit scale. to transform to the prob scale:
$$p=\frac{e^\eta}{1+e^\eta}.$$

```r
prob.bands = exp(se.bands)/(1+exp(se.bands))
```

```r
matplot(age.grid, prob.bands, col = 'blue', type = 'l', ylim= c(0.0, 0.1))
points(jitter(age), I(wage>250)/10, pch = '|', cex = 0.5)
```

# Splines

```r
require(splines)
```

```r
fit = lm(wage~bs(age, knots = c(25, 40, 60)), data = Wage)
```

```r
plot(age, wage, col = "darkgrey")
lines(age.grid, predict(fit, list(age=age.grid)), col = 'darkred', 
      lwd = 2)
abline(v=c(25, 40, 60), lty=2, col='darkred')
fits = smooth.spline(age, wage, df = 16)
lines(fits, col = 'darkgreen', lwd = 2)
fitl = smooth.spline(age, wage, cv=TRUE)
lines(fitl, col = 'darkblue', lwd = 2)
```
