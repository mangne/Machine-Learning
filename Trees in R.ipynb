```r
obb.err = double(13)
test.err = double(13)
```

```r
for(mtry in 1:13){
    fit = randomForest(medv~., data = Boston , subset = train, mtry = mtry)
    obb.err[mtry] = fit$mse[400]
    pred = predict(fit, Boston[-train, ])
    test.err[mtry] = with(Boston[-train,], mean((medv-pred)^2))
}
```

```r
matplot(1:mtry, cbind(test.err, obb.err), pch = 19, col = c('red', 'blue'),
       type = 'b', ylab = 'Mean Sq Error')
legend("topright", legend = c('Test', "OOB"), col = c("red", "blue"), pch = 19)
```

# Boosting

```r
install.packages('gbm')
require(gbm)
```

```r
boost.boston = gbm(medv~., data = Boston[train, ], distribution= 'gaussian',
                  n.trees = 10000, shrinkage = 0.01, interaction.depth = 4)
```

```r
summary(boost.boston)
```

```r
plot(boost.boston, i = "lstat")
```

```r
plot(boost.boston, i = "rm", col = 'red')
```

```r
n.trees = seq(from =100, to=10000, by = 100)
```

```r
predmat = predict(boost.boston, Boston[-train, ], n.trees = n.trees)
```

```r
dim(predmat)
```

```r
perr = with(Boston[-train, ], apply((predmat-medv)^2, 2, mean))
```

```r
plot(n.trees, perr, pch = 19, ylab = 'Mean Sq Error', xlab = '#trees')
abline(h=11.1, col = 'red', lwd = 2)
```
